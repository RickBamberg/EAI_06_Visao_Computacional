## Fundamentos de OpenCV

### Importação das Bibliotecas

```python
import cv2
from matplotlib import pyplot as plt
```

### Leitura de imagem com OpenCV (BGR)

```python
img_bgr = cv2.imread('lena.png')  # Substitua pelo caminho correto

# Conversão de BGR para RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Exibição com Matplotlib
plt.imshow(img_rgb)
plt.title("Imagem RGB")
plt.axis('off')
plt.show()
```

### 2. Redimensionamento e transformação

```python
img_resize = cv2.resize(img_rgb, (200, 200))
plt.imshow(img_resize)
plt.title("Redimensionada")
plt.axis('off')
plt.show()
```

### 3. Conversão de cores

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.title("Escala de Cinza")
plt.axis('off')
plt.show()
```

### 4. Desenhos e anotações

```python
img_draw = img_rgb.copy()
cv2.rectangle(img_draw, (50, 50), (150, 150), (255, 0, 0), 2)  # Retângulo azul
cv2.putText(img_draw, 'Exemplo', (50, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
plt.imshow(img_draw)
plt.title("Com Anotações")
plt.axis('off')
plt.show()
```

### 5. Filtros e detecção de bordas

```python
img_blur = cv2.GaussianBlur(img_gray, (5, 5), 0)
img_edges = cv2.Canny(img_gray, 100, 200)

plt.subplot(1, 2, 1)
plt.imshow(img_blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_edges, cmap='gray')
plt.title("Detecção de Bordas")
plt.axis('off')
plt.show()
```

### 6. Recorte de regiões (ROI)

```python
roi = img_rgb[50:150, 50:150]  # Recorte da área (y1:y2, x1:x2)
plt.imshow(roi)
plt.title("Região de Interesse (ROI)")
plt.axis('off')
plt.show()
```

### 7. Pré-processamento para OCR (binarização)

```python
_, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img_thresh, cmap='gray')
plt.title("Imagem Binarizada")
plt.axis('off')
plt.show()
```
