### üßæ Introdu√ß√£o
A detec√ß√£o de bordas e a aplica√ß√£o de filtros s√£o etapas fundamentais no processamento de imagens. Elas permitem:

- Reduzir ru√≠dos

- Real√ßar contornos

- Melhorar resultados de segmenta√ß√£o e detec√ß√£o

O OpenCV fornece diversas fun√ß√µes para aplicar filtros e detectar bordas de forma eficiente.

### üß™ Importa√ß√µes e Leitura da Imagem

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leitura da imagem em escala de cinza
img = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.show()
```

### Filtros de Suaviza√ß√£o (Blur)
- M√©dia

- Gaussiano

- Mediana

```python
def adicionar_ruido_sal_pimenta(imagem, proporcao_salpimenta=0.05):
    img_ruidosa = imagem.copy()
    num_pixels = imagem.size
    num_salpimenta = int(proporcao_salpimenta * num_pixels)
    
    # Adicionar pixels ‚Äúsal‚Äù (branco)
    coords_s = [np.random.randint(0, i - 1, num_salpimenta) for i in imagem.shape]
    img_ruidosa[coords_s[0], coords_s[1]] = 255
    
    # Adicionar pixels ‚Äúpimenta‚Äù (preto)
    coords_p = [np.random.randint(0, i - 1, num_salpimenta) for i in imagem.shape]
    img_ruidosa[coords_p[0], coords_p[1]] = 0
    
    return img_ruidosa

# Leitura da imagem em escala de cinza
img = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)

# Adicionar ru√≠do sal e pimenta
img_ruidosa = adicionar_ruido_sal_pimenta(img, proporcao_salpimenta=0.05)

# Aplicar filtros
blur = cv2.blur(img_ruidosa, (5, 5))
gaussian = cv2.GaussianBlur(img_ruidosa, (5, 5), 0)
median = cv2.medianBlur(img_ruidosa, 5)

# Mostrar imagens lado a lado
titulos = ['Original com ru√≠do', 'M√©dia', 'Gaussiano', 'Mediana']
imagens = [img_ruidosa, blur, gaussian, median]

plt.figure(figsize=(12, 4))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(imagens[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')
plt.show()
```

### Filtro Bilateral

```python
bilateral = cv2.bilateralFilter(img, 9, 75, 75)

plt.imshow(bilateral, cmap='gray')
plt.title('Filtro Bilateral')
plt.axis('off')
plt.show()
```

### 5. Detec√ß√£o de Bordas com Filtros
### 5.1 Filtro Laplaciano

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
plt.imshow(np.abs(laplacian), cmap='gray')
plt.title('Laplaciano')
plt.axis('off')
plt.show()
```

### 5.2 Filtro Sobel (X e Y)

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(np.abs(sobelx), cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(np.abs(sobely), cmap='gray')
plt.title('Sobel Y')
plt.axis('off')
plt.show()
```

### Detec√ß√£o de Bordas com Canny

```python
edges = cv2.Canny(img, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title('Canny')
plt.axis('off')
plt.show()
```
