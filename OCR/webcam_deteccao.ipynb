## üìò Detec√ß√£o de Imagem pela Webcam
- Objetivo: Utilizar YOLOv5 para realizar detec√ß√£o de objetos em tempo real usando a webcam.

### üîπ Importa√ß√µes Iniciais

```python
import torch
import cv2
import numpy as np
```

### üîπ Carregar o modelo YOLOv5

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', trust_repo=True)
model.eval()
```

### üîπ Captura da Webcam

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Erro ao abrir a webcam.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Converte BGR para RGB (YOLO usa RGB)
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Realiza a infer√™ncia
    results = model(img_rgb)

    # Converte resultado para formato OpenCV
    img_with_boxes = np.squeeze(results.render())  # retorna imagem BGR

    # Exibe
    cv2.imshow('Detec√ß√£o com YOLOv5', img_with_boxes)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libera recursos
cap.release()
cv2.destroyAllWindows()
```
